```python
import pandas as pd

# Read the CSV with dtype specified if you know the types; otherwise, read a small sample to inspect types
df_project = pd.read_csv("Rail_Equipment_Accident_Incident_Data__Form_54_.csv", skipinitialspace=True)

# Assuming you want to strip whitespace from string columns
for col in df_project.columns:
    if df_project[col].dtype == 'object':
        df_project[col] = df_project[col].str.strip()

# Check the DataFrame's information again
df_project.info()
```

Dropping any columns with more 10% NaN values

```python

# Create a list to hold columns to drop
columns_to_drop = []

# Calculate the threshold for dropping columns
threshold = len(df_project) *0.1

# Identify columns that meet the criteria for dropping
for col in df_project.columns:
    if df_project[col].isnull().sum() > threshold:
        columns_to_drop.append(col)

# Drop the identified columns
df_project.drop(columns=columns_to_drop, axis=1, inplace=True)

# Print the DataFrame info
df_project.info()
```

```python


# Select numerical columns, those with data types of int or float
numerical_cols = df_project.select_dtypes(include=['int64', 'float64']).columns

categorical_cols = df_project.select_dtypes(include=['object', 'category']).columns

# Create DataFrames containing just the numerical and categorical columns
df_numerical = df_project[numerical_cols]
df_categorical = df_project[categorical_cols]

df_numerical.drop(['Accident Year','Accident Month', 'Maintenance Accident Year', 'Maintenance Accident Month', 'Day','Year'], axis=1, inplace=True)

df_categorical.drop(['PDF Link', 'State Abbreviation', 'Accident Number', 'Reporting Railroad Code','Maintenance Accident Number','Maintenance Railroad Code','Primary Accident Cause','Accident Cause Code','Accident Cause','Narrative','Incident Key','Report Key','Reporting Parent Railroad Code','Maintenance Parent Railroad Code','Reporting Railroad Holding Company','Class Code'], axis=1, inplace=True)
```

```python
print(df_numerical.info())
```

```python
#change Data to DateTime format
df_categorical['Date'] = pd.to_datetime(df_categorical['Date'], format='%d-%m-%Y', errors='coerce')

print(df_categorical.info())
```

```python
import pandas as pd

def find_mixed_type_columns(df):
    mixed_type_columns = []
    for col in df.columns:
        # Get the set of different types in each column
        types = set(df[col].apply(type))
        # If there's more than one type, add to the list
        if len(types) > 1:
            mixed_type_columns.append(col)
    return mixed_type_columns

# Load your data
df = pd.read_csv("Rail_Equipment_Accident_Incident_Data__Form_54_.csv", low_memory=False)

# Find mixed type columns
mixed_columns = find_mixed_type_columns(df)

# Create a new DataFrame with only the mixed type columns
mixed_columns_df = df[mixed_columns]

# Now mixed_columns_df contains only the columns with mixed types
print(mixed_columns_df.info())
```
